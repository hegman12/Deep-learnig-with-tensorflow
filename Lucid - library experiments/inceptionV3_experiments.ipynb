```python
import numpy as np
import tensorflow as tf

import lucid.modelzoo.vision_models as models
from lucid.misc.io import show
import lucid.optvis.objectives as objectives
import lucid.optvis.param as param
import lucid.optvis.render as render
import lucid.optvis.transform as transform
```

```python
v3=models.inceptionV3()
v3.load_graphdef()
```

```python
l2=lambda layer: objectives.L2(layer=layer)
sp=lambda layer,w: objectives.sigmoid_prob(layer=layer,w=w)
tv=lambda layer: objectives.total_variation(layer=layer)
dd=lambda layer: objectives.deepdream(layer=layer)
lg=lambda layer,l: objectives.class_logit(layer=layer,label=l)
```

```python
obj=l2('InceptionV3/InceptionV3/Conv2d_1a_3x3/convolution')
_=render.render_vis(model=v3,objective_f=obj)
```

```python
obj=sp('InceptionV3/InceptionV3/Conv2d_1a_3x3/Relu',1)
_=render.render_vis(model=v3,objective_f=obj)
```

```python
obj=sp('InceptionV3/InceptionV3/Mixed_7c/Branch_2/Conv2d_0b_3x3/convolution',1000.0)
_=render.render_vis(model=v3,objective_f=obj)
```

```python
obj=dd('InceptionV3/InceptionV3/Mixed_7c/Branch_2/Conv2d_0b_3x3/convolution')
_=render.render_vis(model=v3,objective_f=obj)
```
