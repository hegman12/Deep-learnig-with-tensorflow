```python
from PIL import Image
def save_resized_images():
    
    dirlist = [ os.path.join("C:\\MLDatabases\\google-images-download-master\\downloads\\100x100_resized_animals", item) for item in os.listdir("C:\\MLDatabases\\google-images-download-master\\downloads\\100x100_resized_animals")]
    
    size=1367+1540
    image_array=np.zeros((size,100,100,3))
    
    start_index=0
    y_array=np.zeros(size)
    
    for ims in dirlist:  
        
        image=misc.imread(ims,mode="RGB")
        last_dim=image.shape[-1]

        image_array[start_index]=image

        y_array[start_index]=0
        start_index += 1
    
    dirlist = [ os.path.join("C:\\MLDatabases\\google-images-download-master\\downloads\\100x100_resized_plants", item) for item in os.listdir("C:\\MLDatabases\\google-images-download-master\\downloads\\100x100_resized_plants")]
    
    #start_index=len(image_array)
    for ims in dirlist:
        image=misc.imread(ims,mode="RGB")
        image_array[start_index]=image
            
        y_array[start_index]=1
        start_index += 1
    np.save("C:\\MLDatabases\\google-images-download-master\\downloads\\x_non_alpha_composed",image_array)
    np.save("C:\\MLDatabases\\google-images-download-master\\downloads\\y_non_alpha_composite1",y_array)     
```
